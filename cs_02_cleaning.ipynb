```python
import pandas as pd
import numpy as np

BASE      = '/content/drive/MyDrive/customer-segmentation'
RAW       = f'{BASE}/data/raw'
PROCESSED = f'{BASE}/data/processed'

import os
os.makedirs(PROCESSED, exist_ok=True)

print("Ready!")
```

```python
df = pd.read_excel(f'{RAW}/online_retail_II.xlsx', sheet_name='Year 2010-2011')
print(f"Starting shape: {df.shape}")
```

```python
print(f"Before: {df.shape}")

df = df.dropna(subset=['Customer ID'])

print(f"After removing null Customer IDs: {df.shape}")
print(f"Rows removed: {541910 - len(df):,}")
```

```python
print(f"Before: {df.shape}")

df = df.drop_duplicates()

print(f"After removing duplicates: {df.shape}")
print(f"Rows removed: {406830 - len(df):,}")
```

```python
print(f"Before: {df.shape}")

df = df[~df['Invoice'].astype(str).str.startswith('C')]

print(f"After removing cancellations: {df.shape}")
print(f"Rows removed: {401605 - len(df):,}")
```

```python
print(f"Before: {df.shape}")

df = df[df['Quantity'] > 0]

print(f"After removing negative/zero quantities: {df.shape}")
print(f"Rows removed: {392733 - len(df):,}")
```

```python
print(f"Before: {df.shape}")

df = df[df['Price'] > 0]

print(f"After removing zero/negative prices: {df.shape}")
print(f"Rows removed: {392733 - len(df):,}")
```

```python
print(f"Before: {df.shape}")

df = df.dropna(subset=['Description'])

print(f"After removing missing descriptions: {df.shape}")
print(f"Rows removed: {392693 - len(df):,}")
```

```python
df['Customer ID'] = df['Customer ID'].astype(int)

print("Customer ID type:", df['Customer ID'].dtype)
print("Sample values:", df['Customer ID'].head().tolist())
```

```python
print("=== CLEANING SUMMARY ===")
print(f"Original rows:  541,910")
print(f"Final rows:     {len(df):,}")
print(f"Total removed:  {541910 - len(df):,}")
print(f"Data retained:  {len(df)/541910*100:.1f}%")

print("\n=== FINAL DATA TYPES ===")
print(df.dtypes)

print("\n=== MISSING VALUES ===")
print(df.isnull().sum())

print("\n=== SAMPLE CLEAN DATA ===")
df.head()
```

```python
df.to_csv(f'{PROCESSED}/online_retail_clean.csv', index=False)
print("Clean data saved!")
print(f"Location: {PROCESSED}/online_retail_clean.csv")
```
